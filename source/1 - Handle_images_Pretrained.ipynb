```python
import os
import random
import numpy as np
import cv2
```

```python
#Blur Image
def blur_image(img):
    # Blur the image using GaussianBlur
    blurred_image = cv2.GaussianBlur(img, (13, 13), 0)
    return blurred_image

def pixelate_image(img, block_size=4):
    return cv2.resize(cv2.resize(img, (img.shape[1] // block_size, img.shape[0] // block_size),
                      interpolation = cv2.INTER_NEAREST), (img.shape[1], img.shape[0]),
                      interpolation = cv2.INTER_NEAREST)

#ADD Gaussian Noise
def add_gaussian_noise(img):
    mean = 0
    var = 1000
    sigma = var ** 0.5
    gaussian = np.random.normal(mean, sigma, img.shape)
    noisy_image = img + gaussian
    cv2.normalize(noisy_image, noisy_image, 0, 255, cv2.NORM_MINMAX, dtype=-1)
    noisy_image = noisy_image.astype(np.uint8)
    return noisy_image

#Decrease Intensity
def decrease_intensity(image):
    scale = 0.75
    image = image * scale
    image = np.clip(image, 0, 255)
    image = image.astype(np.uint8)
    return image

#Increase Intensity
def increase_intensity(image):
    scale = 1.25
    image = image * scale
    image = np.clip(image, 0, 255)
    image = image.astype(np.uint8)
    return image

def random_effect(image):
    effects = [blur_image, pixelate_image, add_gaussian_noise]
    random_effect = random.choice(effects)
    return random_effect(image)

def random_intensity(image):
    effects = [increase_intensity, decrease_intensity]
    random_effect = random.choice(effects)
    return random_effect(image)

# folder = "/content/drive/MyDrive/DS201_DeepLearning_VNTF_22/dataset/all_raw_data"
folder = "/content/drive/MyDrive/DS201_DeepLearning_VNTF_22/Test_data/all_raw_data"
for filename in os.listdir(folder):

    if 'modified_' not in filename:

      image = cv2.imread(os.path.join(folder, filename))
      height, width = image.shape[:2]
      if height > 300 and width > 200:
          image = random_effect(image)
          image = random_intensity(image)
      else:
          image = random_intensity(image)
      cv2.imwrite(os.path.join(folder, "modified_" + filename), image)
```
