```python
import os
import glob
import pandas as pd

%cd /content/drive/MyDrive/DS201_DeepLearning_VNTF_22/dataset/
# %cd /content/drive/MyDrive/DS201_DeepLearning_VNTF_22/Test_data/
```

```python
name_file = []
for name in glob.glob('all_raw_data/*'):
    name_file.append(name)
    
df = pd.DataFrame(name_file)
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df
```

```python
for i in range(df.shape[0]):
  df[0].loc[i] = df[0].loc[i].strip('all_raw_data/')
```

```python
df
```

```python
count = 0
for i in range(df.shape[0]):
  zero = ''
  count +=1
  name_img = df[0].loc[i].strip('.png')

  old_name = 'all_raw_data/' + name_img + '.png'
  old_xml = 'xml/' + name_img + '.xml'
  old_yolo = 'yolo/' + name_img + '.txt'
  if count < 1000:
    zero += '0'
    if count < 100:
      zero += '0'
      if count < 10:
        zero += '0'
  new_name = 'all_raw_data/temp_' + zero + str(count) + '_temp.png'
  new_xml = 'xml/temp_' + zero + str(count) + '_temp.xml'
  new_yolo = 'yolo/temp_' + zero + str(count) + '_temp.txt'
  os.rename(old_name, new_name)
  try: 
    os.rename(old_xml, new_xml)
    os.rename(old_yolo, new_yolo)
  except:
    print(zero + str(count) + 'khong co nhan')

for name in glob.glob('all_raw_data/*'):
  name_img = name.strip('_temp.png')
  name_img = name_img.strip('all_raw_data/temp_')

  old_name = 'all_raw_data/temp_' + name_img + '_temp.png'
  old_xml = 'xml/temp_' + name_img + '_temp.xml'
  old_yolo = 'yolo/temp_' + name_img + '_temp.txt'

  new_name = 'all_raw_data/img_' + name_img + '.png'
  new_xml = 'xml/img_' + name_img + '.xml'
  new_yolo = 'yolo/img_' + name_img + '.txt'
  
  os.rename(old_name, new_name)
  try:
    os.rename(old_xml, new_xml)
    os.rename(old_yolo, new_yolo)
  except:
    print(name_img + 'khong co nhan')
```
